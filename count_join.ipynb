```python
import pandas as pd
import numpy as np
d= {'rpn': ['pn1', 'pn2', 'pn3', 'pn1', 'pn4', 'pn5', 'pn6', 'pn5'],
'ipn': ['pn2', 'pn3', 'pn1', 'pn4', 'pn2', 'pn6', 'pn5', 'pn7']}
df = pd.DataFrame(data=d, columns=['rpn', 'ipn'])
df
```

```python
temp_df = df.groupby('rpn').count()
temp_df = temp_df.reset_index()
temp_df = temp_df.rename(columns={'ipn':'count'})
temp_df
```

```python
df = pd.merge(df, temp_df, on='rpn')
```
