```python
import pandas as pd
d= {'rpn': ['pn1', 'pn2', 'pn3', 'pn1', 'pn4', 'pn5', 'pn6', 'pn5'],
'ipn': ['pn2', 'pn3', 'pn1', 'pn4', 'pn2', 'pn6', 'pn5', 'pn7']}
l = {}
def for_loop(ll):
    for i in ll:
        l[i] = ll
def mapped(x):
    if x[0] in l.keys():
        ll = l[x[0]]
        ll.add(x[1])
        l[x[0]] = ll
        for_loop(ll)
    else:
        ll = {(x[0])}
        ll.add(x[1])
        l[x[0]] = ll
        for_loop(ll)
    if x[1] in l.keys():
        ll = l[x[1]]
        ll.add(x[0])
        l[x[1]] = ll
        for_loop(ll)
    else:
        ll = {(x[1])}
        ll.add(x[0])
        l[x[1]] = ll
        for_loop(ll)
    return ll
df = pd.DataFrame(data=d, columns=['rpn', 'ipn'])
df['related_pn'] = df.apply(mapped, axis=1)
df
```
